<b> 1.Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name,experience_required. You have to scrape first 10 jobs data</b>

```python
driver.get("https://www.naukri.com/data-analyst-jobs-in-bangalore-bengaluru?k=data analyst&l=bangalore/bengaluru")



#finding element for job search bar

#search_job = driver.find_element_by_id('qsb-keyword-sugg')

#search_job.send_keys("Data Scientist")

#search_loc = driver.find_element_by_xpath("//input[@name = 'location']")

#search_loc.send_keys("Banglore")



#link_click = driver.find_element_by_xpath("//button[@class ='btn']")

#link_click.click()



jt = driver.find_elements_by_xpath("//a[@class = 'title fw500 ellipsis']")

job_title = []

for i in jt:

    job_title.append(i.text)

    

exp = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")

Experience = []

for i in exp:

    Experience.append(i.text)

    

sal = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi salary']/span")

Salary = []

for i in sal:

    Salary.append(i.text)

    

loc = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")

Location = []

for i in loc:

    Location.append(i.text)
df=pd.DataFrame({})
df['Job_Title']=job_title[:11]
df['Experience']=Experience[:11]
df['Salary']=Salary[:11]
df['Location']=Location[:11]  
df
```

<b>6.Scrape data of first 100 sunglasses listings on flipkart.com. You have to
